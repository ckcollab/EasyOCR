```python
import numpy as np


# Get labels..
training_labels = np.loadtxt('all_data/en_train_filtered/labels.csv', dtype=str, delimiter=',', skiprows=1, usecols=1)
validation_labels = np.loadtxt('all_data/en_val/labels.csv', dtype=str, delimiter=',', skiprows=1, usecols=1)
```

```python
# Turn all the labels into individual characters in a long list..
training_characters = []
for s in training_labels:
    for c in s.replace(' ', ''):
        training_characters.append(c)
validation_characters = []
for s in validation_labels:
    for c in s.replace(' ', ''):
        validation_characters.append(c)
        

# print(training_characters)
```

```python
# Histogram for training...
import pandas
from collections import Counter


df = pandas.DataFrame.from_dict(Counter(training_characters), orient='index').sort_index()
df.plot(kind='bar', fontsize=18, rot=0)
```

```python
df = pandas.DataFrame.from_dict(Counter(validation_characters), orient='index').sort_index()
df.plot(kind='bar', fontsize=18, rot=0)
```
